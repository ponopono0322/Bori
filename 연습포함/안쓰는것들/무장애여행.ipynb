```python
import os
import pandas as pd
Dir = "C:/Users/dwshi/Downloads/제주/csv파일/무장애여행/"
FileList = os.listdir(Dir)
```

```python
Ldata = pd.DataFrame(columns=['위치','위도','경도','장소명칭'])

for i in FileList:
    global Ldata
    data = pd.read_csv(Dir + i, encoding='cp949')
    data['위치'] = i[16:-4]
    data.rename(columns={'lat':'위도', 'lon':'경도', 'title':'장소명칭', 'Title':'장소명칭'}, inplace=True)
    data = data[['위치','위도','경도','장소명칭']]
    Ldata = pd.concat([Ldata, data])
    
Ldata
```

```python
Ldata = Ldata.dropna()
```

```python
import pandas as pd
from pandas import Series, DataFrame
import folium
from folium.plugins import MarkerCluster
map1 = folium.Map(location=[33.389548, 126.239267],zoom_start=15)
mc = MarkerCluster().add_to(map1)

for idx, row in Ldata.iterrows():
    loc = [row[1],row[2]]
    folium.Marker(loc, popup=row[3]).add_to(mc)

map1
```
