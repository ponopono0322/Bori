```python
import pandas as pd

df = pd.read_excel('./크롤링 데이터/수정된파일.xlsx')
df
```

```python
df = df[['타입','장소','위치','해시태그']]
df
```

```python
def to_list(arr):
    ml = arr.replace(' ','').split("'")
    if '[' in ml:
        ml.remove('[')
    if ']' in ml:
        ml.remove(']')
    while ',' in ml:
        ml.remove(',')
    return ml
```

```python
df['해시태그'] = df.apply(lambda val:to_list(val.해시태그), axis=1)
```

```python
view_df = df[df['타입'] == '관광지']
tag_view = []
for x in view_df['해시태그']:
    for y in x:
        if y not in tag_view:
            tag_view.append(y)
print(tag_view)
print(len(tag_view))
```

```python
count=0
for x in df['해시태그']:
    if '사계절' in x:
        count+=1
print(count)
```

```python
who_with = ['부모','친구','커플','혼자','아이']
weather = ['비,눈','흐림','맑음']
taste = ['경관/포토','휴식','테마공원','문화유적지','미술/박물관',
         '액티비티','체험관광','드라이브','쇼핑']
season = ['봄','여름','가을','겨울','사계절']

def per_maker(item, get_arr):
    if item in get_arr:
        return round(1/len(get_arr), 4)
    else:
        return 0

def cate_maker(df_arr, set_arr):
    dic = {}
    intersection = list(set(df_arr) & set(set_arr))
    for i in set_arr:
        dic[i] = per_maker(i, intersection)
    return dic

view_df['일행'] = view_df.apply(lambda val:cate_maker(val.해시태그, who_with), axis=1)

view_df = view_df.set_index('장소')

view_df['날씨'] = view_df.apply(lambda val:cate_maker(val.해시태그, weather), axis=1)
view_df['취향'] = view_df.apply(lambda val:cate_maker(val.해시태그, taste), axis=1)
view_df['계절'] = view_df.apply(lambda val:cate_maker(val.해시태그, season), axis=1)
```
