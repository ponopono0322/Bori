```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
data = pd.read_csv('모두포함.csv', encoding='utf-8')
data.drop('Unnamed: 0', axis=1, inplace=True)
data.dropna(inplace=True)
#data.set_index('장소', inplace=True)
data
```

```python
feature_cols = ['부모님', '아이', '친구', '커플', '친척','가족','지인','연인','혼자','둘','셋','넷']
X = data[feature_cols]
y = data.장소
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print("훈련 세트 정확도: {:.3f}".format(clf.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(clf.score(X_test, y_test)))
```

```python
#정규화해보기
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(data)
data_scaled = scaler.transform(data)
data_df_scaled = pd.DataFrame(data=data_scaled, columns=data.columns, index=data.index)
data_df_scaled
```

```python
finder = ['연인','둘','드라이브','산책']
f_data = data_df_scaled.sort_values(by=finder, ascending=False)
f_data
```

```python
pure_name = f_data[finder].mean(axis=1).sort_values(ascending=False).head(10).index.to_list()
pure_name
```

```python
all_data = pd.read_excel('./크롤링 데이터/수정된파일.xlsx')
all_data.drop('Column1', axis=1, inplace=True)
all_data
```

```python
only_tour = all_data[all_data['타입']=='관광지']
only_tour = only_tour.loc[:,'타입':'기타 상세']
only_tour
```

```python
pure_df = pd.DataFrame()

for name in pure_name:   
    df = only_tour[only_tour['장소'].str.contains(name)]
    pure_df = pure_df.append(df)
pure_df
```

```python
pure_df.to_csv('top_10.csv', encoding='utf-8')
```

```python
def to_list(arr):
    ml = arr.replace(' ','').split("'")
    if '[' in ml:
        ml.remove('[')
    if ']' in ml:
        ml.remove(']')
    while ',' in ml:
        ml.remove(',')
    return ml

pure_df['음식점'] = pure_df.apply(lambda val:to_list(val.음식점), axis=1)
pure_df['숙박'] = pure_df.apply(lambda val:to_list(val.숙박), axis=1)
pure_df.set_index('숙박', inplace=True)
pure_df.reset_index(inplace=True)
```

```python
room_list = pure_df['숙박'][1]
room_list
```

```python
category_df = pd.read_csv('./크롤링 데이터/category_data.csv')
category_df.drop('Unnamed: 0', axis=1, inplace=True)
category_df
```

```python
room_df = pd.DataFrame()

for name in room_list:
    df = category_df[category_df['name'] == name]
    room_df = room_df.append(df)
room_df
```
